# Analysis of Meetup.com Performance

This is an analysis of the benefit of the Meetup.com group for Huntsville AI.

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('https://raw.githubusercontent.com/HSV-AI/presentations/master/2021/meetup-members.csv')
df["Joined"] = pd.to_datetime(df["Joined"])
df["Visited"] = pd.to_datetime(df["Visited"])

grouped = df.groupby(df.Joined).count()
```

# Group Membership over Time

It looks like the number of members has been increasing very well over time. You can even figure out when our NeurIPS meetups occured, since there was usually a slight increase during December.

```python
grouped.cumsum()["Member"].plot(figsize=(10,8), title="Members by Date")
plt.show()
```

# Last Visit Date

Below is a graph that shows the count of members based on their last visit date for the group. I believe this is only captured for visiting the group page on Meetup.com - not an actual meeting attendance.

Looking at the raw data, many members never returned to the page after initially joining the group.

```python
df.groupby(df.Visited).count()["Attended"].plot(figsize=(10,8), title="Last Visited")
plt.show()
```

# Members Grouped by Attendence Count

Below is a graph showing the number of users based on the number of times they attended a meetup. Again, this assumes that a member RSVP actually attended.

```python
grouped = df.groupby(df.Attended).count()["Visited"]
df["Attended"].plot.hist(bins=len(grouped), figsize=(10,8), title="Count by Attendance")
plt.show()
print(grouped)
```

# Results

It is pretty easy to see that most people that find the group and join through Meetup.com do not participate in the meetups.

The top 15 members ranked by attendance are all on a first name basis with the organizers and each other. There really is no benefit achieved by using Meetup.com to organize our events.
